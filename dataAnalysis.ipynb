### 오른쪽 눈

(43~47길이 + 44~46길이)

```python
A = pow(df.iloc[:, 73] - df.iloc[:, 81], 2) + pow(df.iloc[:, 74] - df.iloc[:, 82], 2)
B = pow(df.iloc[:, 75] - df.iloc[:, 79], 2) + pow(df.iloc[:, 76] - df.iloc[:, 80], 2)
C = pow(df.iloc[:, 85] - df.iloc[:, 93], 2) + pow(df.iloc[:, 86] - df.iloc[:, 94], 2)
D = pow(df.iloc[:, 87] - df.iloc[:, 91], 2) + pow(df.iloc[:, 88] - df.iloc[:, 92], 2)
```

```python
distLeft = (np.sqrt(A)+np.sqrt(B))/2
distRight = (np.sqrt(C)+np.sqrt(D))/2
```

```python
plt.figure(figsize=(20, 8))
plt.ylim(0, 15)
plt.xlim(75, 120)

plt.plot(distLeft)
plt.hlines(distLeft.mean(), 0, 160, color='r')
plt.plot(distRight)
plt.hlines(distRight.mean(), 0, 160, color='g')

plt.show()
```

```python
dfWork = pd.read_csv('work.csv', header=None, index_col = 0)
dfWeb = pd.read_csv('web.csv', header=None, index_col = 0)
```

```python
dfWeb.head()
```

```python
plt.figure(figsize=(20, 8))

plt.plot(dfWork)
plt.plot(dfWeb)

plt.show()
```

```python
A = pow(dfWork.iloc[:, 73] - dfWork.iloc[:, 81], 2) + pow(dfWork.iloc[:, 74] - dfWork.iloc[:, 82], 2)
B = pow(dfWork.iloc[:, 75] - dfWork.iloc[:, 79], 2) + pow(dfWork.iloc[:, 76] - dfWork.iloc[:, 80], 2)
C = pow(dfWork.iloc[:, 85] - dfWork.iloc[:, 93], 2) + pow(dfWork.iloc[:, 86] - dfWork.iloc[:, 94], 2)
D = pow(dfWork.iloc[:, 87] - dfWork.iloc[:, 91], 2) + pow(dfWork.iloc[:, 88] - dfWork.iloc[:, 92], 2)
distLeft = (np.sqrt(A)+np.sqrt(B))/2
distRight = (np.sqrt(C)+np.sqrt(D))/2
```

```python
plt.figure(figsize=(20, 8))
plt.ylim(0, 15)
#plt.xlim(75, 120)

plt.plot(distLeft)
plt.hlines(distLeft.mean(), 0, 160, color='r')
plt.plot(distRight)
plt.hlines(distRight.mean(), 0, 160, color='g')

plt.show()
```

```python
A = pow(dfWeb.iloc[:, 73] - dfWeb.iloc[:, 81], 2) + pow(dfWeb.iloc[:, 74] - dfWeb.iloc[:, 82], 2)
B = pow(dfWeb.iloc[:, 75] - dfWeb.iloc[:, 79], 2) + pow(dfWeb.iloc[:, 76] - dfWeb.iloc[:, 80], 2)
C = pow(dfWeb.iloc[:, 85] - dfWeb.iloc[:, 93], 2) + pow(dfWeb.iloc[:, 86] - dfWeb.iloc[:, 94], 2)
D = pow(dfWeb.iloc[:, 87] - dfWeb.iloc[:, 91], 2) + pow(dfWeb.iloc[:, 88] - dfWeb.iloc[:, 92], 2)
distLeftWeb = (np.sqrt(A)+np.sqrt(B))/2
distRightWeb = (np.sqrt(C)+np.sqrt(D))/2
```

```python
plt.figure(figsize=(25, 8))
plt.ylim(0, 15)
plt.xlim(0, 240)

plt.plot(distLeft)
plt.plot(distLeftWeb)

plt.show()
```
